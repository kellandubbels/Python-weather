```
# complete script to retrieve basic forecast

# retrieve forecast for FGF/118,31
import urllib.request
forecast = urllib.request.urlopen("https://api.weather.gov/gridpoints/FGF/118,31/forecast").read()
# print(forecast)

# transform byte-like object 'forecast' into string using 'decode()' (using encoding utf-8)
forecast_string = forecast.decode("utf-8")
# print(forecast_string)

# Use RegEx to select rows, outputs a list 'basic_forecast'
# match from properties:...periods:...number >> name/temperature/detailedForecast
import re
regex = re.compile(r"name.*|temperature\".*|detailedForecast.*")
basic_forecast = re.findall(regex, forecast_string)

# print each element of the list 'basic_forecast' with a newline for a separator
print(*basic_forecast, sep = "\n")

print("\n")
three_day_forecast = basic_forecast[0:18]
print(*three_day_forecast, sep = "\n")


# basic_forecast_string = " ".join(basic_forecast)
# print(basic_forecast_string)
# split_forecast_string = basic_forecast_string.split(".")
# print(split_forecast_string)
```
